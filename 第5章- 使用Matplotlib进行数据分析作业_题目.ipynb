```python

GDP = pd.read_csv('世界各国GDP（1960-2019）.csv', encoding='gbk')
GDP.drop(columns='2020', axis=1, inplace=True)
GDP.head()
```

### Step11 根据国家真实GDP数据画图：统计1965年所有国家中NaN的数值，并绘制直方图

```python
gdp_1965 = GDP['1965']
nan_count = gdp_1965.isna().sum()
non_nan_count = gdp_1965.count()
print(gdp_1965)
print(nan_count)
print(non_nan_count)
```

```python
# your code

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.bar(['NaN', 'No-NaN'], [nan_count, non_nan_count])
plt.ylabel('数量')
plt.yticks(range(20, max(nan_count, non_nan_count) + 20, 20))
plt.show()
```

### Step12 根据国家真实GDP数据画图：画出中国1960年至2015年的GDP曲线图，取值可以用切片(iloc[0,4:60])

```python
# your code
from matplotlib.ticker import FuncFormatter
china_gdp = GDP.iloc[40, 4:60] # 这个切片的意思是取第一行，第4列到第60列的数据
years = range(1960, 2016)
plt.figure(figsize=(10, 6))
plt.plot(years, china_gdp / 1e13, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('GDP (1e13)')
plt.title('中国GDP')
plt.grid(True)

def scientific_notation(x, pos):
    return f"{x:.0e}"

formatter = FuncFormatter(scientific_notation)
plt.gca().yaxis.set_major_formatter(formatter)
# plt.show()
# print(china_gdp)
```

### Step13 根据国家真实GDP数据画图：画出以所有国家2015年GDP为小于百亿，小于千亿，小于万亿和大于万亿的国家占比的饼图，可以将NaN填为0

```python
# your code
plt.rcParams['font.sans-serif'] = ['SimHei']
GDP.fillna(0, inplace=True)
gdp_2015 = GDP['2015']
less_than_100b = gdp_2015[gdp_2015 < 1e11].count()
less_than_1t = gdp_2015[(gdp_2015 >= 1e11) & (gdp_2015 < 1e12)].count()
less_than_10t = gdp_2015[(gdp_2015 >= 1e12) & (gdp_2015 < 1e13)].count()
greater_than_10t = gdp_2015[gdp_2015 >= 1e13].count()
labels = ['小于百亿', '小于千亿', '小于万亿', '大于万亿']
sizes = [less_than_100b, less_than_1t, less_than_10t, greater_than_10t]
colors = ['lightcoral', 'lightblue', 'lightgreen', 'gold']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=0) 
plt.title('GDP分布')

plt.show()
```

### Step14 根据国家真实GDP数据画图：画出日本与韩国的折线图，两条折线画在一张图中

```python
#画出日本与韩国的折线图，两条折线画在一张图中
# your code
japan_gdp = GDP.iloc[119, 4:60]
korea_gdp = GDP.iloc[126, 4:60]
years = range(1960, 2016)
plt.figure(figsize=(10, 6))
plt.plot(years, japan_gdp , marker='o', linestyle='-', color='b', label='日本')
plt.plot(years, korea_gdp, marker='o', linestyle='-', color='g', label='韩国')
plt.xlabel('Year')
plt.ylabel('GDP (1e12)')
plt.title('日本与韩国GDP')
plt.grid(True)
plt.legend()
plt.show()

```

### Step15 根据国家真实GDP数据画图：画出日本与韩国的面积图，画在一张图中

```python

japan_gdp = GDP.iloc[119, 4:60]
korea_gdp = GDP.iloc[126, 4:60]

japan_percent_change = (japan_gdp / japan_gdp.iloc[0] - 1) * 100
korea_percent_change = (korea_gdp / korea_gdp.iloc[0] - 1) * 100
years = range(1960, 2020)

plt.figure(figsize=(10, 6))
japan_percent_change.plot(kind='area', label='Japan', color='b', alpha=0.5)
korea_percent_change.plot(kind='area', label='South Korea', color='g', alpha=0.5)
plt.xlabel('Year')
plt.grid(True)
plt.legend()
plt.show()
```
